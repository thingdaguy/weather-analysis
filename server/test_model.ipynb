```python
import joblib
import pandas as pd

scaler = joblib.load("models/weather_scaler.joblib")
kmeans = joblib.load("models/weather_kmeans.joblib")

cluster_map = {
    0: "Wet",
    1: "Normal",
    2: "Dry"
}

tests = [
    {"t_max": 28, "t_min": 24, "wind_speed": 18, "rain": 15},  # Wet
    {"t_max": 35, "t_min": 25, "wind_speed": 6,  "rain": 2},   # Normal
    { "t_max": 25, "t_min": 18, "wind_speed": 8,  "rain": 0},   # Dry
]

df_test = pd.DataFrame(tests)
X_scaled = scaler.transform(df_test)
clusters = kmeans.predict(X_scaled)

df_test["cluster"] = clusters
df_test["condition"] = df_test["cluster"].map(cluster_map)

df_test
```
